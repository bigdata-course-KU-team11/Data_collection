```python
for i in range(2): #  range(len(bridge_name))
    driver = webdriver.Chrome()
    url = 'http://hrfco.go.kr/popup/waterlevelPopup.do?Obscd=1013655&Sdt=2020-08-11%2022:00&Edt=2020-08-11%2022:00&Obsnm=' + quote_bn[i]
    driver.get(url)
    driver.find_element_by_xpath('//li[@id="radio2"]').click()
    for j in range(len(list_datestart)):
        driver.find_element_by_xpath('//input[@id="datestart"]').clear()
        driver.find_element_by_xpath('//input[@id="dateend"]').clear()
        driver.find_element_by_xpath('//input[@id="datestart"]').send_keys(list_datestart[j])
        driver.find_element_by_xpath('//input[@id="dateend"]').send_keys(list_dateend[j])
        driver.find_element_by_xpath('//button[@class="serch"]').click()
        driver.find_element_by_xpath('//button[@class="exel"]').click()
    driver.close()
```
